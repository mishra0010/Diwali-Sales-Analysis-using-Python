```python
# Importing the pandas library for data manipulation and analysis
import pandas as pd
# Importing the numpy library for numerical operations, such as array and matrix computations
import numpy as np
# Importing the seaborn library for data visualization, which is built on top of matplotlib
import seaborn as sns
# Importing the pyplot module from matplotlib, used for creating static, interactive, and animated visualizations
import matplotlib.pyplot as plt

```

```python
# Reading the 'Diwali Sales Data.csv' file into a DataFrame using pandas
# The 'encoding' parameter is set to 'unicode_escape' to handle any special characters in the file
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
# To check the number of rows and columns in the DataFrame
df.shape
```

```python
# To view the first 5 rows of the DataFrame to get a quick overview of the data
df.head()
```

```python
# To get a summary of the DataFrame, including the data types of each column, non-null counts, and memory usage
df.info()
```

```python
# To remove the 'Status' and 'unnamed1' columns from the DataFrame
# The 'axis=1' parameter indicates that columns are being dropped (axis=0 would drop rows)
# The 'inplace=True' parameter modifies the existing DataFrame without creating a new one
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# To check for any missing (null) values in each column of the DataFrame
# The result shows the count of null values in each column
pd.isnull(df).sum()
```

```python
df.shape
```

```python
# To remove any rows with missing (null) values from the DataFrame
# The 'inplace=True' parameter modifies the existing DataFrame directly, removing rows with null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# To convert the 'Amount' column in the DataFrame to integer data type
# This ensures that the 'Amount' column contains integer values rather than other data types
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
# To generate descriptive statistics for the DataFrame's numeric columns
# This includes measures such as count, mean, standard deviation, min, max, and percentiles
df.describe()
```

# Exploratory Data Analysis

### Genders

```python
# To create a count plot of the 'Gender' column using seaborn
# This visualizes the frequency of each gender category in the DataFrame
ax1 = sns.countplot(x='Gender', data=df)

# To add labels on top of each bar in the count plot for better readability
# The 'ax1.containers' contains all the bar containers, and 'ax1.bar_label(bars)' adds labels
for bars in ax1.containers:
    ax1.bar_label(bars)
```

```python
# To group the DataFrame by the 'Gender' column and calculate the total 'Amount' for each gender
# The 'as_index=False' parameter ensures that 'Gender' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# To create a bar plot of the total 'Amount' for each 'Gender'
# The 'x' parameter specifies the categorical variable, and 'y' specifies the numeric variable
ax2 = sns.barplot(x='Gender', y='Amount', data=sales_gen)

# To add labels on top of each bar in the bar plot for better readability
# The 'ax2.containers' contains all the bar containers, and 'ax2.bar_label(bars)' adds labels
for bars in ax2.containers:
    ax2.bar_label(bars)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

# Age

```python
# To create a count plot of 'Age Group' with different colors for each 'Gender'
# The 'hue' parameter adds a gender-based color distinction in the plot
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

# To add labels on top of each bar in the count plot for better readability
# The 'ax.containers' contains all the bar containers, and 'ax.bar_label(bars)' adds labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To group the DataFrame by 'Age Group' and calculate the total 'Amount' for each age group
# The 'as_index=False' parameter ensures that 'Age Group' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# To create a bar plot of the total 'Amount' for each 'Age Group'
# The 'x' parameter specifies the categorical variable, and 'y' specifies the numeric variable
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

# State

```python
# To group the DataFrame by 'State' and calculate the total number of 'Orders' for each state
# The 'as_index=False' parameter ensures that 'State' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Orders' in descending order, and only the top 10 states are selected
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# To set the size of the plot to be wider for better visualization
sns.set(rc={'figure.figsize':(15,5)})

# To create a bar plot of the total number of 'Orders' for the top 10 states
# The 'x' parameter specifies the categorical variable (state), and 'y' specifies the numeric variable (orders)
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
# To group the DataFrame by 'State' and calculate the total 'Amount' for each state
# The 'as_index=False' parameter ensures that 'State' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order, and only the top 10 states are selected
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# To set the size of the plot to be wider for better visualization
sns.set(rc={'figure.figsize':(15,5)})

# To create a bar plot of the total 'Amount' (sales) for the top 10 states
# The 'x' parameter specifies the categorical variable (state), and 'y' specifies the numeric variable (amount)
sns.barplot(data=sales_state, x='State', y='Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


# Marital Status 

```python
# To create a count plot of 'Marital_Status' to visualize the frequency of each marital status category
ax = sns.countplot(data=df, x='Marital_Status')

# To set the size of the plot to be smaller and more compact
sns.set(rc={'figure.figsize':(7,5)})

# To add labels on top of each bar in the count plot for better readability
# The 'ax.containers' contains all the bar containers, and 'ax.bar_label(bars)' adds labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To group the DataFrame by 'Marital_Status' and 'Gender' and calculate the total 'Amount' for each combination
# The 'as_index=False' parameter ensures that both 'Marital_Status' and 'Gender' remain columns in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# To set the size of the plot to be slightly smaller for better visualization
sns.set(rc={'figure.figsize':(6,5)})

# To create a bar plot of the total 'Amount' for each 'Marital_Status', with different colors for each 'Gender'
# The 'hue' parameter adds a gender-based color distinction in the plot
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are unmarried (women) and they have high purchasing power*

# Occupation 

```python
# To set the size of the plot to be much wider for better visualization of the 'Occupation' distribution
sns.set(rc={'figure.figsize':(20,5)})

# To create a count plot of 'Occupation' to visualize the frequency of each occupation category
ax = sns.countplot(data=df, x='Occupation')

# To add labels on top of each bar in the count plot for better readability
# The 'ax.containers' contains all the bar containers, and 'ax.bar_label(bars)' adds labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To group the DataFrame by 'Occupation' and calculate the total 'Amount' for each occupation
# The 'as_index=False' parameter ensures that 'Occupation' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# To set the size of the plot to be much wider for better visualization of the 'Occupation' sales data
sns.set(rc={'figure.figsize':(20,5)})

# To create a bar plot of the total 'Amount' for each 'Occupation'
# The 'x' parameter specifies the categorical variable (occupation), and 'y' specifies the numeric variable (amount)
sns.barplot(data=sales_state, x='Occupation', y='Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

# Product Category

```python
# To set the size of the plot to be very wide for better visualization of the 'Product_Category' distribution
sns.set(rc={'figure.figsize':(25,5)})

# To create a count plot of 'Product_Category' to visualize the frequency of each product category
ax = sns.countplot(data=df, x='Product_Category')

# To add labels on top of each bar in the count plot for better readability
# The 'ax.containers' contains all the bar containers, and 'ax.bar_label(bars)' adds labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# To group the DataFrame by 'Product_Category' and calculate the total 'Amount' for each category
# The 'as_index=False' parameter ensures that 'Product_Category' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Amount' in descending order, and only the top 10 categories are selected
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# To set the size of the plot to be wider for better visualization of the top 10 product categories
sns.set(rc={'figure.figsize':(20,5)})

# To create a bar plot of the total 'Amount' for the top 10 'Product_Category'
# The 'x' parameter specifies the categorical variable (product category), and 'y' specifies the numeric variable (amount)
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
# To group the DataFrame by 'Product_ID' and calculate the total number of 'Orders' for each product
# The 'as_index=False' parameter ensures that 'Product_ID' remains a column in the resulting DataFrame
# The resulting DataFrame is sorted by 'Orders' in descending order, and only the top 10 products are selected
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# To set the size of the plot to be wider for better visualization of the top 10 products by order count
sns.set(rc={'figure.figsize':(20,5)})

# To create a bar plot of the total number of 'Orders' for the top 10 'Product_ID's
# The 'x' parameter specifies the categorical variable (product ID), and 'y' specifies the numeric variable (orders)
sns.barplot(data=sales_state, x='Product_ID', y='Orders')
```
